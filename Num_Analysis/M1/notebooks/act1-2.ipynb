## Raíces de Polinómios. 

```python
# Dependencias y funciones:
import numpy as np

def div_sintetica_lineal(coeficientes, c):
    """
    Divide un polinomio P(x) entre (x - c) por división sintética.
    Retorna los coeficientes del cociente y el residuo.
    """
    q = []
    for a in coeficientes:
        if not q:
            q.append(a)
        else:
            q.append(a + q[-1] * c)
    return q[:-1], q[-1]

def div_sintetica_cuadratica(coeficientes, b, c):
    """
    Divide un polinomio P(x) entre un factor cuadrático (x^2 + bx + c).
    Retorna los coeficientes del cociente y el residuo [Rx, S].
    """
    out = list(coeficientes)
    for i in range(len(coeficientes) - 2):
        coef = out[i]
        if coef != 0:
            out[i + 1] -= coef * b
            out[i + 2] -= coef * c
    return out[:-2], out[-2:]

def imprimir_resultado(cociente, residuo, es_cuadratica=False):
    """
    Función auxiliar para imprimir los resultados de forma limpia.
    (Sugerida por IA).
    """
    print(f"Cociente: {np.poly1d(cociente)}")
    if es_cuadratica:
        print(f"Residuo:  {np.poly1d(residuo)}")
    else:
        print(f"Residuo:  {residuo}")
```

### Pregunta 1
Utilizando división sintética divida el polinomio:
$$
p(x) = x^4 - 5x^3 + 5x^2 + 5x - 6
$$, 
por el factor $x - 2$. ¿Es $x = 2$ una raíz de $p$?

```python
# Coeficientes de p(x)
p1 = [1, -5, 5, 5, -6]
raiz_1 = 2

cociente_1, residuo_1 = div_sintetica_lineal(p1, raiz_1)

imprimir_resultado(cociente_1, residuo_1)
if residuo_1 == 0:
    print("Como el residuo es 0, x = 2 SÍ es una raíz de p(x).")
else:
    print("Como el residuo es diferente de cero, x = 2 No es una raíz de p(x).")
```

### Pregunta 5
Considere el polinomio:
$$
p(x) = x^5 + 4x^4 - 375x^3 + 2510x^2 - 856x - 14784
$$.
Utilizando el procedimiento de división sintética y sabiendo que las raíces del polinomio son $\{-2, 4, 7, 11, -24\}$, compruebe que la derivada de $p(x)$ evaluada en cada raíz es la misma que la evaluación del cociente que queda al utilizar cada raíz respectivamente.

```python
p5_coefs = [1, 4, -375, 2510, -856, -14784]
raices_5 = [-2, 4, 7, 11, -24]

# Construimos el polinomio con numpy y sacamos su derivada
p5_poly = np.poly1d(p5_coefs)
p5_der = np.polyder(p5_poly)

print(f"Polinomio p(x):\n{p5_poly}\n")
print(f"Derivada p'(x):\n{p5_der}\n")
print("-" * 50)
print(f"{'Raíz':>5} | {'p\'(raíz)':>12} | {'Q(raíz)':>12} | {'¿Son iguales?':>15}")
print("-" * 50)

for r in raices_5:
    # 1. División sintética para obtener el cociente Q(x)
    q_coefs, res = div_sintetica_lineal(p5_coefs, r)
    q_poly = np.poly1d(q_coefs)
    
    # 2. Evaluamos la derivada original en 'r'
    eval_der = p5_der(r)
    
    # 3. Evaluamos el cociente resultante en 'r'
    eval_q = q_poly(r)
    
    # Comprobación
    son_iguales = "Sí" if eval_der == eval_q else "No"
    
    print(f"{r:5} | {eval_der:12} | {eval_q:12} | {son_iguales:>15}")
```
