#### Señales y Sistemas en Ciencia de Datos

##### **Definición de Señal**

Una **señal** es un conjunto de puntos de datos que contienen información dentro de un intervalo de una variable independiente (tiempo, espacio, etc.).

Matemáticamente, una señal puede modelarse como:

$$
x : \mathcal{D} \rightarrow \mathbb{R}
$$

donde:

- $\mathcal{D}$ es el dominio de la variable independiente.  
- Cada medición corresponde a un punto independiente del dominio.  
- Existe una correspondencia inyectiva entre la información y su representación funcional.

> **Idea clave:** Una señal es una representación matemática de información percibida y registrada.

---

##### **Clasificación de las Señales**

Las señales se clasifican según la naturaleza de sus variables.

###### **1. Señal Continua (Analógica)**

- Variable independiente continua.  
- Variable dependiente continua.

$$
x(t), \quad t \in \mathbb{R}
$$

---

###### **2. Señal *Piece-wise***

- Variable independiente continua.  
- Variable dependiente discreta por intervalos.

$$
x(t) \in \{v_1, v_2, \dots, v_k\}, \quad t \in \mathbb{R}
$$

---

###### **3. Señal Discreta**

- Variable independiente discreta.  
- Variable dependiente continua o discreta.

$$
x[n], \quad n \in \mathbb{Z}
$$

---

###### **4. Señal Digital**

- Ambas variables continuas.  
- No existe relación entre valores consecutivos de la variable dependiente.

$$
x(t) \in \mathbb{R}, \quad t \in \mathbb{R}
$$

---

#### Sistemas

##### **Definición de Sistema**

Un **sistema** es una relación entre señales donde existe una señal de entrada (*input*) y una señal de salida (*output*).

Puede modelarse como un operador sobre señales:

$$
\sigma : C_1 \rightarrow C_2
$$

donde:

- $C_1$ es un conjunto de señales de entrada del tipo $x(t)$.  
- $\sigma$ es el operador (sistema).  
- $C_2$ es el conjunto de señales de salida del tipo $y(t)$.

La transformación se expresa como:

$$
y(t) = \sigma\{x(t)\}
$$

> **Idea clave:** Un sistema transforma señales de entrada en señales de salida mediante una regla definida.


##### **Sistema Lineal**

Un sistema es **lineal** si cumple la propiedad de **superposición**:

$$
\sigma\left[\alpha x_1(t) + \beta x_2(t)\right] = \alpha \, \sigma\left[x_1(t)\right] + \beta \, \sigma\left[x_2(t)\right]
$$

donde:

- $\alpha, \beta \in \mathbb{R}$ son constantes.
- $x_1(t), x_2(t)$ son señales de entrada.

> **Idea clave:** Si la combinación lineal de entradas produce la misma combinación lineal en las salidas, entonces $\sigma$ es un sistema lineal.

---

##### **Sistema Causal**

Un sistema es **causal** si su salida en el instante actual depende únicamente de valores presentes o pasados de la entrada, **nunca de valores futuros**.

> **Interpretación:** Podemos computar el estado actual usando solo información ya disponible.

###### **Ejemplo de sistema causal**

$$
x(t) \xrightarrow{\ \sigma\ } x(t-1)
$$

- En $t=0$: $x(0) \rightarrow x(-1)$  
- En $t=1$: $x(1) \rightarrow x(0)$  

Este sistema es causal porque solo utiliza valores pasados de la señal.

---

###### **Ejemplo de sistema no causal**

$$
x(t) \xrightarrow{\ \sigma\ } \frac{x(t+1) + x(t)}{2}
$$

- En $t=0$:

$$
\frac{x(1) + x(0)}{2}
$$

No se conoce aún $x(1)$, por lo que el sistema depende de un valor futuro.

> **Conclusión:** El sistema no es causal.

---

##### **Sistema Invariante en el Tiempo**

Un sistema es **invariante en el tiempo** si un desplazamiento (delay) en la entrada produce el mismo desplazamiento en la salida.

Sistema original:

$$
x(t) \xrightarrow{\ \sigma\ } y(t)
$$

Aplicando un desplazamiento $t_0$ en la entrada:

$$
x(t - t_0) \xrightarrow{\ \sigma\ } x\big(2(t - t_0)\big)
$$

Si desplazamos directamente la salida:

$$
y(t - t_0) = x\big(2(t - t_0)\big)
$$

> **Idea clave:** El sistema es invariante porque el efecto del desplazamiento es consistente entre entrada y salida.


#### **Sistema LTI (Linear Time-Invariant)**

Un sistema **LTI** es aquel que cumple simultáneamente:

- **Linealidad**
- **Invariancia en el tiempo**

(LTI = *Linear Time-Invariant*)

---

##### **Propiedad Fundamental**

La característica central de los sistemas LTI es que **su salida ante cualquier entrada puede obtenerse mediante una convolución**.

---

##### **Descomposición de una Señal (Tiempo Discreto)**

Una señal discreta puede representarse como una suma ponderada de impulsos desplazados:

$$
x[n] = \sum_{k=-\infty}^{\infty} x[k] \, \delta[n-k]
$$

donde:

- $\delta[n-k]$ es el impulso unitario desplazado.
- $x[k]$ actúa como el peso (coeficiente) de cada impulso.

> **Idea clave:** Toda señal discreta puede construirse como una combinación lineal de impulsos desplazados.

---

##### **Definición de Convolución (Tiempo Discreto)**

Si el sistema es LTI y $h[n]$ es su respuesta al impulso, entonces la salida está dada por:

$$
y[n] = \sum_{k=-\infty}^{\infty} x[k] \, h[n-k]
$$

---

##### **Definición de Convolución (Tiempo Continuo)**

En el caso continuo, la salida se define como:

$$
y(t) = \int_{-\infty}^{\infty} x(\tau) \, h(t-\tau) \, d\tau
$$

---

#### **Interpretación Conceptual de la Convolución**

Para calcular la respuesta del sistema ante una señal de entrada:

1. **Reflejamos y desplazamos** la respuesta al impulso $h$ a lo largo del tiempo.
2. En cada instante, **multiplicamos** el valor de la entrada $x[k]$ por la respuesta del sistema en ese desplazamiento.
3. **Sumamos (o integramos)** todos los productos para obtener la salida en ese instante.

> **Intuición:** La salida es la acumulación ponderada de versiones desplazadas de la respuesta al impulso.
